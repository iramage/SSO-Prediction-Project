```python
# Import packages

import numpy as np
import pandas as pd
import requests as re
import time
from datetime import datetime
from dateutil.relativedelta import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV -- parsing dates takes ~5 minutes

Sdf = pd.read_csv('Sdf_Sept7.csv', parse_dates=['START_DATE'], index_col=0)
```

```python
SSdf = Sdf.groupby('SPILL_ID').first()

```

```python
SSdf = SSdf.reset_index()

print(SSdf.info())
SSdf
```

```python
indexes = list(SSdf.index)
batch = indexes[0:7000]
```

```python
# Download from Dark Sky

secret_code = '71c36490e95e70a0d9e1c10074360cb7'  # for DarkSky API

Rdf = pd.DataFrame()  # make empty Rain data frame

for i in batch : 
    
    sid = SSdf.iloc[i, 0]  # get Spill ID
    print("Spill #" + str(i) + ": " + str(sid))   

    date = SSdf.iloc[i, 1]   ## have to use local time (START_DATE)
    d = date.strftime("%Y-%m-%dT%H:%M:%S")
    d1 = date - relativedelta(days=1)
    d_minus1 = d1.strftime("%Y-%m-%dT%H:%M:%S")
    d2 = date - relativedelta(days=2)
    d_minus2 = d2.strftime("%Y-%m-%dT%H:%M:%S")   
    d3 = date - relativedelta(days=3)
    d_minus3 = d3.strftime("%Y-%m-%dT%H:%M:%S") 
    d4 = date - relativedelta(days=4)
    d_minus4 = d4.strftime("%Y-%m-%dT%H:%M:%S") 
    dates = [d_minus4, d_minus3, d_minus2, d_minus1, d]  # need day of event plus four days prior
    print(dates)

    lat = SSdf.iloc[i, 2]
    lon = SSdf.iloc[i, 3]
    
    DD = []
    fivedays = pd.DataFrame()  # make empty df for the five days of Dark Sky data (will combine to make Rdf)
    
    for date in dates :
        r = re.get('https://api.darksky.net/forecast/' + secret_code + '/' + str(lat) + ',' + str(lon) + ',' + date  
            + '?exclude=currently,minutely,daily&units=si')  # just hourly (returns data for 24-hr day around time given)        
        try:
            DD.append(r.json()) 
        except:         
            print("Error....")

    j = 0
    
    for j in range(5) :  
    
        data = pd.DataFrame.from_dict(DD[j]['hourly']['data']) 
    
        data['datetime'] = data['time']
        for k in range(len(data['time'])) :
            data['datetime'][k] = datetime.fromtimestamp(data['time'][k])
            
        data['DTNS'] = DD[j]['flags']['nearest-station']  # distance to nearest recording station

        fivedays = pd.concat([fivedays,data])

#    sns.lineplot(x=fivedays['datetime'], y=fivedays['temperature'])
    
    fivedays['SPILL_ID'] = sid   # set Spill ID to merge with SSdf later
    fivedays['START_DATE'] = SSdf.iloc[i, 1]  # preserve time of spill
    
    Rdf = pd.concat([Rdf,fivedays])
                    
```

```python
# Make time a UTC date

Rdf['UTC_time'] = pd.to_datetime(Rdf['time'], unit='s')
```
