```python
# Import packages

import pandas as pd
```

```python
RS = pd.read_csv('Rain-Caused Spills with Less Info.csv', parse_dates=['START_DATE'])
print(RS.info())
```

```python
SMAP = pd.read_csv('Spills with (Old) SMAP Data.csv')
print(SMAP.info())
```

```python
SMAP = SMAP.drop_duplicates(subset='SPILL_ID', keep='first', inplace=False)
print(SMAP.info())
```

```python
SMAP = SMAP.sort_values('SPILL_ID')
SMAP
```

```python
RSMAP = RS.merge(SMAP, on='SPILL_ID')
RSMAP = RSMAP.sort_values('SPILL_ID')
RSMAP
```

```python
RSMAP = RSMAP.drop_duplicates(subset='SPILL_ID', keep='first', inplace=False)
print(RSMAP.info())
```

```python
RSMAP.to_csv('I&I Spills with SMAP Data.csv')
```
