```python
# Import packages

import pandas as pd
```

```python
NSAD_over10mm = pd.read_csv('NSAD_10mm (1494 entries).csv', parse_dates=['START_DATE'])
NSAD_over10mm
```

```python
NSAD_over10mm = NSAD_over10mm.sort_values('4-Day Rain')
NSAD_over10mm = NSAD_over10mm.sort_values('SPILL_CAUSE')
print(NSAD_over10mm [NSAD_over10mm['SPILL_CAUSE'].str.contains('FOG', case=True)] )

```

```python
# Remove FOG and Wipes/Non-Dispersables if Rain < 20mm

NSAD_remove = NSAD_over10mm [ (NSAD_over10mm['SPILL_CAUSE'].str.contains('FOG', case=True) == True) | 
                              (NSAD_over10mm['SPILL CAUSE EXPLAIN'].str.contains('FOG', case=True) == True) |
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Wipes', case=False) == True) | 
                              (NSAD_over10mm['SPILL CAUSE EXPLAIN'].str.contains('Wipes', case=False) == True) |         
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Dispersables', case=False) == True) | 
                              (NSAD_over10mm['SPILL CAUSE EXPLAIN'].str.contains('Dispersables', case=False) == True) ]
NSAD_remove
```

```python
NSAD_remove = NSAD_remove.sort_values('4-Day Rain')

print(NSAD_remove [NSAD_remove['4-Day Rain'] > 30] )

```

```python
discards = list(NSAD_remove['SPILL_ID'])
NSAD_over10mm = NSAD_over10mm [ ~NSAD_over10mm['SPILL_ID'].isin(discards) ]
NSAD_over10mm
```

```python
NSAD_over10mm.to_csv('NSAD_10mm (1300 entries).csv')
```

```python
NSAD_remove = NSAD_over10mm [ (NSAD_over10mm['SPILL_CAUSE'].str.contains('Air Relief Valve', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('CS Maintenance Caused', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Construction Diversion Failure', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Damage by Others', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Debris from Construction', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Operator Error', case=False) == True) | 
                              (NSAD_over10mm['SPILL_CAUSE'].str.contains('Vandalism', case=False) == True) ]
NSAD_remove
```

```python
discards = list(NSAD_remove['SPILL_ID'])
NSAD_over10mm = NSAD_over10mm [ ~NSAD_over10mm['SPILL_ID'].isin(discards) ]
NSAD_over10mm
```

```python
NSAD_over10mm.to_csv('NSAD_10mm (1221 entries).csv')
```

```python
SASD_under10mm = pd.read_csv('SA-SD Spills with Under 10mm Rain (Keep 102).csv', parse_dates=['START_DATE'])
SASD_under10mm
```

```python
SASD = pd.read_csv('SA-SD Spills (1015 entries).csv', parse_dates=['START_DATE'], index_col=0)
SASD
```

```python
# Jinane's search phrase captured "drain" so remove

SASD = SASD [ SASD['SPILL_CAUSE'].str.contains('drain', case=False) == False ] 
SASD = SASD [ SASD['SPILL CAUSE EXPLAIN'].str.contains('drain', case=False) == False ] 
SASD
```

```python
SASD = SASD.sort_values('SPILL_ID')
SASD
```

```python
SASD.to_csv("SA-SD Spills (1015 entries).csv")
```

```python
# Add 4-Day Rain to SASD

four_day = pd.read_csv('4day_cumulative.csv')
four_day = four_day.rename(columns={'precipIntensity':'4-Day Rain'})
four_day
```

```python
SASD_withrain = SASD.merge(four_day, on='SPILL_ID')
SASD_withrain
```

```python
SASD_withrain = SASD_withrain.sort_values('SPILL_ID')
SASD_withrain
```

```python
spills_withrain = list(SASD_withrain['SPILL_ID'])
SASD_norain = SASD[ ~SASD['SPILL_ID'].isin(spills_withrain) ]

SASD_norain
```

```python
SASD_norain = SASD_norain.sort_values('SPILL_ID')
SASD_norain
```

```python
SAD = pd.concat([SASD_withrain,SASD_norain,SASD_under10mm]).reset_index(drop=True)
SAD
```

```python
SAD = SAD.sort_values('SPILL_ID')
SAD
```

```python
All = pd.concat([SAD,NSAD_over10mm],sort=True).reset_index(drop=True)
All

# Got 2246 entries
```

```python
All = All.sort_values('SPILL_ID')
All
```

```python
print(All.info())
```

```python
print(All.isna().sum())

# only missing 425 rain values
```

```python
All.to_csv('Rain-Caused Spills.csv')

```
