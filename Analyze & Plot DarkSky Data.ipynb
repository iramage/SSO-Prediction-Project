```python
# Import packages

import numpy as np
import pandas as pd
import datetime

```

```python
# Read DarkSky data

DS1 = pd.read_csv("Rdf[0-6999]_Sept12.csv", parse_dates=['START_DATE', 'datetime', 'UTC_time'], index_col=0)
print(DS1.info())

```

```python
# Read Part 2

DS2 = pd.read_csv("Rdf[7000-15082]_Sept12.csv", parse_dates=['START_DATE', 'datetime', 'UTC_time'], index_col=0)

```

```python
DS = pd.concat([DS1,DS2],axis=0)
print(DS.info())
```

```python
DS.to_csv('DarkSky_Data_Sept13.csv')
```

```python
DSS = DS.groupby('SPILL_ID').first()
print(DSS)
```

```python
D = pd.concat([DSS['START_DATE'],DSS['DTNS']], axis=1) 
D
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
D.plot(subplots=True)
plt.show()

```

```python
D['DTNS'].plot(kind='hist')
plt.show()

```

```python
plt.hist(D['DTNS'])
plt.show()
```

```python
plt.hist(D['DTNS'], bins='auto')
plt.ylabel("No. of Spills")
plt.xlabel("Distance to Nearest Station (km)")

plt.savefig("DTNS Histogram.jpg")
plt.show()
```

```python
D = D.sort_values('DTNS')
D
```

```python
print(D.tail(20))
```

```python
# Put in buckets

D_under1km = D[ D['DTNS'] < 1 ]
D_1to2km = D[ (D['DTNS'] >= 1) & (D['DTNS'] < 2) ]
D_2to3km = D[ (D['DTNS'] >= 2) & (D['DTNS'] < 3) ]
D_over3km = D[ D['DTNS'] > 3 ]

```

```python
# Create pie chart

counts = [len(D_under1km), len(D_1to2km), len(D_2to3km), len(D_over3km)]
labels = ['Less than 1km', '1km to 2km', '2km to 3km', 'Greater than 3km']
plt.pie(counts, labels=labels, startangle=90, autopct='%.1f%%')
plt.title('Distances to Stations')
plt.savefig("Distances to Stations.jpg")

plt.show()
```

```python
#  Create nicer pie chart


counts = [len(D_under1km), len(D_1to2km), len(D_2to3km), len(D_over3km)]
labels = ['Less than 1km', '1km to 2km', '2km to 3km', 'Greater than 3km']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(counts, labels=labels, explode=explode, startangle=0, colors=colors, autopct='%.1f%%', shadow=True)
plt.title('Distances to Stations')
plt.savefig("Distances to Stations.jpg")

plt.show()
```
